# Preprocessing Pipeline

Interactive walkthrough of MIDI parsing, piano-roll conversion, and dataset statistics.

```python
import os, sys, json, glob
import numpy as np
import matplotlib.pyplot as plt
import pretty_midi
sys.path.insert(0, '../src')
```

```python
from preprocessing.midi_parser import midi_to_events, process_dataset
from preprocessing.piano_roll import events_to_piano_roll
from preprocessing.tokenizer import events_to_tokens, tokens_to_events, GENRES
```

```python
# Process dataset
meta = process_dataset('../data/raw_midi', '../data/processed', max_files=100)
print(f'Processed {len(meta)} files')
```

```python
# Visualise one piano roll
sample_files = glob.glob('../data/processed/*.json')
if sample_files:
    with open(sample_files[0]) as f:
        d = json.load(f)
    roll = events_to_piano_roll(d['events'], n_bars=4)
    plt.figure(figsize=(14, 4))
    plt.imshow(roll, aspect='auto', origin='lower', cmap='hot')
    plt.xlabel('Time Steps'); plt.ylabel('Pitch (0=MIDI 21)')
    plt.title(f"Piano Roll - {d['genre']} ({len(d['events'])} notes)")
    plt.colorbar(label='Velocity')
    plt.tight_layout(); plt.show()
```

```python
# Token statistics
if sample_files:
    tokens = events_to_tokens(d['events'], d['genre'])
    print(f'Sequence length : {len(tokens)} tokens')
    print(f'First 20 tokens : {tokens[:20]}')
    reconstructed = tokens_to_events(tokens)
    print(f'Reconstructed events: {len(reconstructed)}')
```
